```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("C:/Users/joelj/Downloads/IMDb Movies India.csv/IMDb Movies India.csv",encoding="latin1")
df.head()#displays first five rows of the DataFrame
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Convert Duration to string and extract numeric values
df['Duration'] = df['Duration'].astype(str).str.extract('(\d+)')[0].astype(float)

# One-hot encoding categorical variables
df = pd.get_dummies(df, columns=['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3'])

# Separating features and target variable
x = df.drop(columns=['Name', 'Year', 'Rating', 'Votes'])
y = df['Rating']

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Filling NaN values with the mean
x_train = x_train.fillna(x_train.mean())
x_test = x_test.fillna(x_test.mean())
y_train = y_train.fillna(y_train.mean())
y_test = y_test.fillna(y_test.mean())

# Scaling the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Now, x_train, x_test, y_train, and y_test are ready for model training
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
model=Ridge(alpha=1.0)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
print("Mean squared error",mse)
```

```python
plt.scatter(y_test,y_pred,color="blue",alpha=0.5)
#plots a red dashed diagonal line from the minimum to the maximum values of y_test, typically used as a reference line to compare predicted versus actual values in a scatter plot.
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],color="red",linestyle="--")
plt.xlabel("Actual Rating")
plt.ylabel("Predicted Rating")
plt.title("Actual vs Predictied Ratings")
plt.show()
```
